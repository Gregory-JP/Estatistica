## Desvio padrão com dados agrupados

```python
import numpy as np
```

```python
dados = {'inferior': [150, 154, 158, 162, 166, 170],
         'superior': [154, 158, 162, 166, 170, 174],
         'fi': [5, 9, 11, 7, 5, 3]}
```

```python
import pandas as pd

df = pd.DataFrame(dados)
df
```

```python
df['xi'] = (df['superior'] + df['inferior']) / 2
df
```

```python
df['fi.xi'] = df['fi'] * df['xi']
df
```

```python
df['Fi'] = 0
df
```

```python
frequencia_acumulada = []
somatorio = 0

for linha in df.iterrows():
    # print(linha[1][2])
    
    somatorio += linha[1][2]
    frequencia_acumulada.append(somatorio)
```

```python
frequencia_acumulada
```

```python
df['Fi'] = frequencia_acumulada
df
```

```python
df['xi^2'] = df['xi'].values ** 2
```

```python
df
```

```python
df['fi.xi^2'] = df['fi'] * df['xi^2']
df
```

```python
df.columns
```

```python
colunas_ordenadas = ['inferior', 'superior', 'fi', 'xi', 'fi.xi', 'xi^2', 'fi.xi^2', 'Fi']
```

```python
df = df[colunas_ordenadas]
df
```

## Cálculo

```python
import math
```

```python
dp = math.sqrt((df['fi.xi^2'].sum() / df['fi'].sum()) - (df['fi.xi'].sum() / df['fi'].sum()) ** 2)
```
