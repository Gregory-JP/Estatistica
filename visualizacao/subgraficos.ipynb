```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../bases_dados/census.csv')
df.head()
```

```python
g = sns.FacetGrid(df, col = 'sex', hue = 'income')
g.map(sns.scatterplot, 'age', 'final-weight');
```

```python
g = sns.FacetGrid(df, col = 'workclass', hue = 'income')
g.map(sns.scatterplot, 'age', 'final-weight');
```

```python
g = sns.FacetGrid(df, col = 'sex', hue = 'income')
g.map(sns.histplot, 'age');
```

```python
df2 = df.iloc[:, [0, 4, 12]]
df2.head()
```

```python
g = sns.PairGrid(df2)
g.map(sns.scatterplot);
```

```python
g = sns.PairGrid(df2)
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot);
```
