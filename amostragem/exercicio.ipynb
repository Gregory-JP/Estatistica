## Amostragem Estratificada

```python
from sklearn.model_selection import StratifiedShuffleSplit
df['c#default'].value_counts()

def amostragem_estratificada(df, percentual, campo):
    split = StratifiedShuffleSplit(test_size=percentual, random_state=1)
    
    for _, y in split.split(df, df[campo]):
        df_y = df.iloc[y]
    return df_y
```

```python
df_amostragem_estratificada = amostragem_estratificada(df, 0.5, 'c#default')
df_amostragem_estratificada.shape
```
