```python
def amostragem_grupos(df, n_grupos):
    intervalo = len(df) / n_grupos
    
    grupos = []
    id_grupo = 0
    contagem = 0

    for _ in df.iterrows():
        grupos.append(id_grupo)
        contagem += 1
    
        if contagem > intervalo:
            contagem = 0
            id_grupo += 1
        
    df['grupo'] = grupos
    random.seed(1)
    grupo_selecionado = random.randint(0, n_grupos)
    return df[df['grupo'] == grupo_selecionado]
```

```python
df_amostra_agrupamento = amostragem_grupos(df, 100)
df_amostra_agrupamento.shape
```
