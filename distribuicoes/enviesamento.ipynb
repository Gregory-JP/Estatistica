```python
import seaborn as sns
from scipy.stats import skewnorm
import numpy as np
import scipy.stats
```

```python
dados_normal = skewnorm.rvs(a=0, size=1000)
```

```python
sns.displot(dados_normal, kde=True);
```

```python
dados_normal.mean()
```

```python
np.median(dados_normal)
```

```python
scipy.stats.mode(dados_normal)
```
